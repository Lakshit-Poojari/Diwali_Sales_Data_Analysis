```python
cp = sea.countplot(x="Gender", data=df, hue="Gender")

for bar in cp.containers:
    cp.bar_label(bar)
```

```python
sales_gend=df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False)
sea.barplot(x="Gender",y="Amount",data=sales_gend,hue="Gender")
```

```python
ax = sea.countplot(x="Age Group", data=df, hue="Gender")

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_gend=df.groupby(["Age Group","Gender"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False)
sea.barplot(x="Age Group",y="Amount",data=sales_gend,hue="Gender")
```

```python
sx=sea.countplot(x="State",data=df, hue="Gender")
sea.set(rc={"figure.figsize":(25,5)})
for bar in sx.containers:
    sx.bar_label(bar)

```

```python
sales_satate=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False)

sea.barplot(x="State",y="Orders", data = sales_satate,hue="State")
```

```python
sales_satate=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False)

sea.barplot(x="State",y="Amount", data = sales_satate,hue="State")
```

```python
zx=sea.countplot(x="Zone",data=df,hue="Gender")
sea.set(rc={"figure.figsize":(5,5)})
for bar in zx.containers:
    zx.bar_label(bar)
```

```python
zone_sales=df.groupby(["Zone","Gender"],as_index=False)["Amount"].sum().sort_values("Amount", ascending=False)
sea.set(rc={"figure.figsize":(10,5)})
sea.barplot(x="Zone",y="Amount",data=zone_sales,hue="Gender")
```
